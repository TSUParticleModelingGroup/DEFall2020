```python
import math
import numpy as np
from matplotlib import pyplot as plt

# Sets the size of the window
plt.figure(figsize=(8,8))

# Setting up the axes we will be drawing on
fMin = -5
fMax = 5
gMin = -5
gMax = 5
plt.axis([fMin, fMax, gMin, gMax])
plt.ylabel('g', fontsize=20)
plt.xlabel('f', fontsize=20)
plt.title('Phase Plane', fontsize=20)

# Place your DEs below.
def Df(f,g):
    "This is the f differintail equation that will generate the DF"
    #return((-1.0*g)*(g-2.0))
    return(f)
    #return(5.0*f-3.0*g-2.0)
    #return(-g*(g-2))
    #return(((0.1*g*f)-(0.05*f)))

def Dg(g,f):
    "This is the g differintail equation that will generate the DG"
    #return((f-2.0)*(g-2.0))
    return(2.0*g)
    #return(4.0*f-3*g-1.0)
    #return((f-2)*(g-2))
    #return(((g-(0.1*g*f))))

# Grid size
gridf = 0.3
gridg = 0.3

# The following code draw the Phase Plane
f = fMin
while f <= fMax:
    g = gMin
    while g <= gMax:
        plt.plot([f], [g], 'ko', ms = 2.0)
        df = Df(f,g)
        dg = Dg(g,f)
        length = math.sqrt(df*df + dg*dg)
        adjustment = (gridf + gridg)/2.0
        try:
            plt.arrow(f, g, (df/length)*adjustment, (dg/length)*adjustment, head_width=0.04, head_length=0.08, fc='b', ec='r')
        
        except:
            pass
        g += gridg
    f += gridf

plt.show()
```
