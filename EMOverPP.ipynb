```python
import math
import numpy as np
from matplotlib import pyplot as plt

# Sets the size of the window
plt.figure(figsize=(8,8))

# Setting up the axes we will be drawing on
fMin = -5
fMax = 5
gMin = -5
gMax = 5
plt.axis([fMin, fMax, gMin, gMax])
plt.ylabel('f', fontsize=20)
plt.xlabel('g', fontsize=20)
plt.title('Phase Plane', fontsize=20)

# Put DEs here
def Df(f,g):
    "This is the f differintail equation that will generate the Df"
    #return(f)
    #return(5.0*f-3.0*g-2.0)
    return(-g*(g-2.0))

def Dg(g,f):
    "This is the g differintail equation that will generate the Dg"
    #return(2.0*g)
    #return(4.0*f-3.0*g-1)
    return((f-2.0)*(g-2.0))

# Grid size
gridf = 0.3
gridg = 0.3

# Drawing the phase plain
f = fMin
while f <= fMax:
    g = gMin
    while g <= gMax:
        plt.plot([f], [g], 'ko', ms = 2.0)
        df = Df(f,g)
        dg = Dg(g,f)
        length = math.sqrt(df*df + dg*dg)
        adjustment = (gridf + gridg)/2.0
        plt.arrow(f, g, (df/length)*adjustment, (dg/length)*adjustment, head_width=0.04, head_length=0.08, fc='b', ec='r')
        g += gridg
    f += gridf

# Setting time values
tStart = -3.0
totalRunTime = 10.0
tStop = tStart + totalRunTime
dt = 0.01

# Initial values of the functions should be evaluated at tStart.
fStart = 1.0
gStart = 1.0

# The following code calculates and draws Euler's method
fOld = fStart
gOld = gStart
t = tStart
plt.plot([fStart], [gStart], 'go', ms = 5.0)

while t <= tStop:
    fNew = fOld + dt*Df(fOld,gOld)
    gNew = gOld + dt*Dg(gOld,fOld)
    lines = plt.plot([fOld, fNew], [gOld,gNew], color='b', linewidth=1.0)
    #print(fNew,gNew) # uncoment this line to see numeric outcomes
    t = t + dt
    fOld = fNew
    gOld = gNew
print("Final Value",t, fNew, gNew)
plt.show()
```
